```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import TensorDataset,DataLoader
import os
import pickle
from utils import * 
```

```python
# 本区块完成数据的加载读取以及切片池化。

# PPG300的每个样本数据特征量为19200，过于大，进行切片采样


#  获取数据集最值，用于Min-max初始化操作
MAXMIN_DICT = {}

# 获取两种数据所有的最值
for TYPE in ["EDA","PPG"]:
    MAX = -10000
    MIN = 10000
    PATH = f'D:\\CrossPC_DS\\{TYPE}\\eight\\'
    for file_name in os.listdir(PATH):
        if(os.path.splitext(file_name)[1] == ".pkl"):
            for row in pd.read_pickle(PATH + file_name)[TYPE].values:
                MAX = np.max(row) if np.max(row) > MAX else MAX
                MIN = np.min(row) if np.min(row) < MIN else MIN
                print(MIN)
#     DICT[TYPE]["MAX"] = MAX
#     DICT[TYPE]["MIN"] = MIN
    MAXMIN_DICT.update({TYPE:{"MAX":MAX,"MIN":MIN}})
    

```

```python
# 超酷炫字典  EDG300
DICT_EDG300 = {}

for TYPE_TARGET in ["PPG","EDA"]:
# for TYPE_TARGET in ["PPG"]:
    # 准备数据集 #PPG
    dataURL = f"D:/CrossPC_DS/{TYPE_TARGET}/eight/{TYPE_TARGET}300.pkl"
    dataX = []
    dataY = []
    data = pd.read_pickle(dataURL).values
    data_dim = len(data[0][4])
    for i in data:
        #dataX.append(i[4][0:1200])
        dataX.append(maxPooling(i[4][0:data_dim],1200))
        dataY.append(i[5])

    dataX = torch.unsqueeze(torch.Tensor(np.stack(dataX)),1)  # reshape至(None,1,1200)
    dataX = (dataX - MAXMIN_DICT[TYPE_TARGET]["MIN"])/( MAXMIN_DICT[TYPE_TARGET]["MAX"] -  MAXMIN_DICT[TYPE_TARGET]["MIN"])# 归一化
    dataY = torch.LongTensor(dataY)  # 分类长整型

    DICT_EDG300.update({TYPE_TARGET:{"dataX":dataX,"dataY":dataY}})
    
DICT_EDG300.update({"EDG":{"dataX":torch.cat((DICT_EDG300["EDA"]["dataX"],DICT_EDG300["PPG"]["dataX"]),dim=1),"dataY":dataY}})

with open('DICT_EDG300.pkl', 'wb') as file:
    pickle.dump(DICT_EDG300, file)
```
